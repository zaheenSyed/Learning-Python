```python
# deque from collections
from collections import deque
dq=deque()
dq.append(1)
dq.append(2)
print(dq)
```

```python
dq.popleft() # FIFO
print(dq)

dq.appendleft(3)
print(dq)

```

```python
# HashSet
mySet = set()

mySet.add(1)
mySet.add(2)
print(mySet)
print(len(mySet))

print(1 in mySet)
print(2 in mySet)
print(3 in mySet)

mySet.remove(2)
print(2 in mySet)

# list to set
print(set([1, 2, 3]))

# Set comprehension
mySet = { i for i in range(5) }
print(mySet)
```
